# 1) Profitability
What types of movies have been the most profitable worldwide?

```python
df1 = pd.read_csv("C:/Users/User/Documents/Phase_1/zipped_data/tn.movie_budgets.csv.gz")
df2 = pd.read_csv("C:/Users/User/Documents/Phase_1/zipped_data/imdb.title.basics.csv.gz")
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3 = df1.reset_index()
```

```python
df4 = df2.reset_index()
```

```python
df3.columns
```

'df3' refers to the csv file - 'tn.movie_budgets'. The information in this dataframe provides insight into the release date, movie title, production budget, domestic gross, and worldwide gross. This data will be useful when understanding the profitability of each movie, domestically and worldwide.

```python
df4.columns
```

'df4' refers to the csv file - 'imdb.title.basics'. The information in this dataframe provides insight into the movie tag, movie title(s), release year, length, and genre. This data will be useful when examining the genre of each movie title.

```python
combined_df = df3.merge(df4, left_on='movie', right_on='primary_title', how='inner')
combined_df
```

Merged 'df3' and 'df4' by the movie names to match up associated information.

```python
combined_df.dropna(inplace=True)
newcombined_df = combined_df['genres'].str.split(",", n=2, expand=True)
combined_df['genre 1']=newcombined_df[0]
combined_df['genre 2']=newcombined_df[1]
combined_df['genre 3']=newcombined_df[2]
combined_df.drop(columns = ['genres'], inplace = True)
combined_df
```

The important information to isolate in this 'combined_df' are the different genres that each movie falls in to. Since the 'genre' column includes multiple genres, it is important to separate them in order to analyze each row specifically. The 'genre' column entries are split and 3 new columns are created, 'genre 1', 'genre 2', 'genre 3', to replace the original 'genre' column.

```python
clean_df = combined_df.drop(columns = ['release_date', 'tconst', 'primary_title', 'original_title', 'start_year', 'runtime_minutes'])
```

```python
clean_df
```

```python
column_values = clean_df[['genre 1', 'genre 2', 'genre 3']].values.ravel()
unique_values = pd.unique(column_values)
unique_values
```

The column values in 'clean_df['genre 1', 'genre 2', 'genre 3']' need to be listed and stated in order to find out how many different genres are listed with the given data set.

```python
genre_1 = clean_df['genre 1'].value_counts()
print(genre_1)
genre_2 = clean_df['genre 2'].value_counts()
print(genre_2)
genre_3 = clean_df['genre 3'].value_counts()
print(genre_3)
```

Now that the different genres have been listed, it's important to understand how many movies fall under each genre category.

```python
data = [genre_1, genre_2, genre_3]
headers = ['genre 1', 'genre 2', 'genre 3']
genre_df = pd.concat(data, axis = 1, keys = headers)
```

```python
genre_df
```

```python
genre_df['total'] = genre_df.sum(axis=1)
print(genre_df)
```

```python
clean_df['production_budget'] = clean_df['production_budget'].map(lambda x: x.lstrip('$').rstrip(','))
clean_df['production_budget'] = clean_df['production_budget'].str.replace(',','')
clean_df['domestic_gross'] = clean_df['domestic_gross'].map(lambda x: x.lstrip('$').rstrip(','))
clean_df['domestic_gross'] = clean_df['domestic_gross'].str.replace(',','')
clean_df['worldwide_gross'] = clean_df['worldwide_gross'].map(lambda x: x.lstrip('$').rstrip(','))
clean_df['worldwide_gross'] = clean_df['worldwide_gross'].str.replace(',','')
```

In order to manipulate the numerical data in the 'production_budget', 'domestic_gross', and 'worldwide_gross' columns, they must be changed from strings to integers. In order to achieve this, unnecessary information such as '$' and commas must be stripped from the column values.

```python
clean_df
```

```python
clean_df['production_budget'] = pd.to_numeric(clean_df['production_budget'])
clean_df['domestic_gross'] = pd.to_numeric(clean_df['domestic_gross'])
clean_df['worldwide_gross'] = pd.to_numeric(clean_df['worldwide_gross'])
```

```python
clean_df
```

```python
clean_df['domestic_profit'] = clean_df['domestic_gross'] - clean_df['production_budget']
clean_df['worldwide_profit'] = clean_df['worldwide_gross'] - clean_df['production_budget']
```

Profit can be calculated by gross earnings minus cost of production. In the case of 'clean_df', 'domestic_gross' and 'worldwide_gross' can be considered the gross earnings and 'production_budget' can be considered the cost of production.

```python
clean_df
```

```python
profit_df = clean_df.drop(['index_x','id','index_y', 'genre 2', 'genre 3'], axis=1)
```

Irrelevant columns can be dropped at this point. 'genre 2' and 'genre 3' were dropped because they are not the primary genres of the associated films.

```python
profit_df
```

```python
profitbygenre_df= profit_df.groupby(profit_df['genre 1']).sum()
print(profitbygenre_df)
```

This 'groupby' was done by 'genre 1' in order to specify the total dollar amounts of associated columns of all films according to their genre type.

```python
ppmbgtemp_df = pd.concat([profitbygenre_df, genre_df['genre 1']], axis=1)
ppmbgtemp_df
```

The two dataframes, 'profitbygenre_df' and 'genre_df', are combined to present the relevant data in order to complete this portion of the analysis. In order to find the average dollar amounts of the associated columns in the 'profitbygenre_df', the number of movies in each genre category must be provided as listed in the 'genre_df'.

```python
ppmbgtemp_df = ppmbgtemp_df.astype(float)
ppmbgtemp_df
```

Before the data in 'ppmbgtemp_df' can be analyzed, the values in the 'profitbygenre_df' and 'genre_df' where 'ppmbgtemp_df's information is derived from must be compatible. Since the column values of 'profitbygenre_df' are 'int' types, they must be converted to the 'float' type in order to comply to the column values of 'genre_df'.

```python
productionbudget_df = ppmbgtemp_df.apply(lambda x: x['production_budget']/x['genre 1'], axis =1)
domesticgross_df = ppmbgtemp_df.apply(lambda x: x['domestic_gross']/x['genre 1'], axis =1)
worldwidegross_df = ppmbgtemp_df.apply(lambda x: x['worldwide_gross']/x['genre 1'], axis =1)
domesticprofit_df = ppmbgtemp_df.apply(lambda x: x['domestic_profit']/x['genre 1'], axis =1)
worldwideprofit_df = ppmbgtemp_df.apply(lambda x: x['worldwide_profit']/x['genre 1'], axis =1)
average_df = pd.concat([productionbudget_df, domesticgross_df, worldwidegross_df, domesticprofit_df, worldwideprofit_df], axis=1)
average_df
```

One of the final step to analyze the data for our intended recommendation is to compute the average dollar values of each column by dividing it by the number of movies in each genre.

```python
xfinalaverage_df = average_df.rename(columns={0: "Avg Production Budget", 1: "Avg Domestic Gross", 2: "Avg Worldwide Gross", 3: "Avg Domestic Profit", 4: "Avg Worldwide Profit"})
```

```python
finalaverage_df = xfinalaverage_df.sort_values(by=['Avg Worldwide Profit'], ascending=False)
finalaverage_df
```

```python
finalaverage_df.reset_index(inplace = True)
```

```python
finalaverage_df.rename(columns={'index': 'Genre'}, inplace=True)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(y = 'Genre', x = 'Avg Worldwide Profit', palette = 'flare', data = finalaverage_df.iloc[0:10])
plt.title('Average Worldwide Profit by Genre')
plt.xlabel('Average Worldwide Profit')
sns.set_style("ticks")
sns.despine()
plt.savefig('profit.png')
```

The final step to this recommendation is to provide a readable and well understood graphical representation of our findings. The pertinent information required are the different movie genres and their associated average worldwide profits.
Additionally, it is important to only provide the pertinent information and remove any irrelevant or unviable considerations from the graphic. Thus, it was determined only the top ten performing genres should be represented and shown.

```python
df1 = pd.read_csv("C:/Users/User/Documents/Phase_1/zipped_data/tmdb.movies.csv.gz")
df2 = pd.read_csv("C:/Users/User/Documents/Phase_1/zipped_data/bom.movie_gross.csv.gz")
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3 = df1.reset_index()
```

```python
df4 = df2.reset_index()
```

```python
df3.columns
```

'df3' refers to the csv file - 'tmdb.movies'. The information in this dataframe provides insight into the genre tag, movie tag, original language, original movie title, popularity, release date, movie title, average vote rating, and vote count. This data will be useful when understanding the specific release date of each movie.

```python
df4.columns
```

'df4' refers to the csv file - 'bom.movie_gross'. The information in this dataframe provides insight into the movie title, production studio, domestic gross, foreign gross, and year of release. This data will be useful when understanding the worldwide gross of each movie.

```python
combined_df = df3.merge(df4, left_on = 'original_title', right_on = 'title', how = 'inner')
combined_df
```

'df3' and 'df4' are merged on the 'original_title' and 'title' respectively in order to match the values of the 'df3' and 'df4' rows accurately.

```python
combined_df.dropna(inplace = True)
newcombined_df = combined_df['release_date'].str.split("-", n=2, expand = True)
combined_df['Year'] = newcombined_df[0]
combined_df['Month Num'] = newcombined_df[1]
combined_df['Day'] = newcombined_df[2]
combined_df.drop(columns = ['index_x', 'Unnamed: 0', 'genre_ids', 'id', 'original_language', 'popularity', 'release_date', 'title_x', 'vote_average', 'vote_count', 'index_y', 'title_y', 'studio', 'year'], inplace = True)
combined_df
```

In order to properly identify the exact date of when each movie was released, the 'release_date' column must be separated to represent the 'year', 'month', and 'day' of release independently. 

```python
combined_df['foreign_gross'] = combined_df['foreign_gross'].replace(',','', regex=True)
combined_df
```

Next, the values types in the 'foreign_gross' columns must be changed from 'str' type to 'float' type in order to properly compute the necessary functions later on. In order to achieve this the data must first be cleaned by removing any unnecessary annotations like commas.

```python
combined_df['foreign_gross'] = combined_df['foreign_gross'].astype(float)
```

```python
combined_df['Worldwide Gross'] = combined_df['domestic_gross'] + combined_df['foreign_gross']
combined_df
```

Since we are examining the worldwide gross of the films, the columns 'domestic_gross' and 'foreign_gross' must be added in order to find the worldwide gross of each film.

```python
combined_df.drop(columns = ['domestic_gross', 'foreign_gross'], inplace = True)
```

Drop the unnecessary information from 'combined_df'.

```python
combined_df
```

```python
combined_df.rename(columns = {'original_title': 'Title'}, inplace = True)
```

```python
combined_df
```

```python
final_df = combined_df.groupby(by = ['Month Num'], dropna = False).mean()
final_df
```

A 'groupby' function is used to group the information by the film's release month. The mean function is called to find what the average worldwide gross is for each month of the year.

```python
final_df.rename(columns = {'Worldwide Gross': 'Average Worldwide Gross'}, inplace = True)
```

The 'Worldwide Gross' column must be renamed to 'Average Worldwide Gross' to correctly represent the information presented.

```python
final_df
```

```python
ffinal_df = final_df.reset_index()
```

```python
ffinal_df
```

```python
ffinal_df['Month'] = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

The months of the year are added in a new column to the dataframe in order to provide a more easily readable representation of the data.

```python
ffinal_df
```

```python
ffinal_df.drop(columns = ['Month Num'], inplace = True)
```

```python
ffinal_df
```

```python
plt.figure(figsize=(15,8))
sns.barplot(y = 'Month', x = 'Average Worldwide Gross', palette = 'flare', data = ffinal_df.iloc[0:12])
plt.title('Average Worldwide Gross by Month')
plt.xlabel('Average Worldwide Gross')
plt.ylabel('Month')
sns.set_style("ticks")
sns.despine()
plt.savefig('time')
```

The final step to this recommendation is to provide a readable and well understood graphical representation of our findings. The pertinent information required are the months in the year and their associated average worldwide gross.
